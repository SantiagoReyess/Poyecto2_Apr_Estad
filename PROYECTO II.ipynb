### REGRESIÓN LOGÍSTICA

Una regresión logística es un modelo matemático de clasificación, que se encarga de calcular la probabilidad de un evento o característica dados un conjunto de factores.



En la regresión logística, se ajusta una curva en forma de “S” a los datos, donde la curva va de 0 a 1 y representa la probabilidad de que según la escala del factor dado, la característica sea real o no. Para encontrar esta curva, se utiliza un método de optimización de máxima verosimilitud, donde buscas adherirte a la mayor cantidad de punto “acertados” posibles. 

### MÁQUINAS DE SOPORTE VECTORIAL (SVM)

Una máquina de soporte vectorial es un algoritmo de aprendizaje estadístico supervisado que se encarga de encontrar una línea o un hiperplano óptimo que maximiza la distancia entre clases en un espacio n-dimensional. Son modelos comúmnente utilizados en problemas de clasificación, ya que es capaz de generalizar de buena manera a nuevos datos a partir de vectores de soporte, los cuales son adyacentes al hiperplano que clasifica los datos. 



[1]

### REDES NEURONALES

Una red neuronal es un programa o modelo de aprendizaje estadístico que se encarga de tomar decisiones, simulando la manera en que las neuronas biológicas del cerebro humano se intercalan y conectan para tomar decisiones. 



Cada nodo en la red neuronal se puede interpretar como un modelo de regresión individual, que captura información diferente a los demás, pero también los nutre conforme se avanza en las capas de la red, para finalmente dar un resultado en una capa de salida.
Es importante mencionar que las redes neuronales no siguen un proceso de optimización convexa, por lo que el máximo valor del método de optimización podría corresponder a un máximo local y no uno global. 


[2]

### KERNELS

Los kernels son funciones matemáticas que ayudan a producir datos a partir de las similaridades de un mismo conjunto, esto permite producir conjuntos de datos no lineales a partir de datos que lo sean. Existen diferentes tipos de kernel, como el polinomial que se encarga de elevar a una potencia dada los datos de un conjunto. El kernel RBF (Radial Basis Function) por ejemplo, se utiliza en las máquinas de soporte vectorial para encontrar la menor distancia entre dos puntos similares, basándose en una distribución gaussiana, que luego sirve para generar los modelos de clasificación. 

### ESPACIOS DE HILBERT

Los espacios de Hilbert son un concepto matemático que se encarga de aplicar o generalizar las reglas geométricas convencionales de un espacio euclideano a uno de dimensiones infinitas. Esto se logra a partir de un espacio vectorial con un producto interno que es completo respecto a la norma de dicho producto. 



Los espacios de Hilbert son importantes para aprendizaje estadístico, ya que se utilizan para transformar los datos que tienen dimensiones difíciles para ser clasificados, y los llevan a un espacio de HIlbert de mayores dimensiones donde si  pueden separarse, sin computar directamente la transformación, utilizando kernels.

### MÉTRICOS PARA CLASIFICACIÓN 

Existen diferentes métricos para evaluar el funcionamiento de los modelos de clasificación, y así verificar su precisión y efectividad. Entre ellos están los siguientes:

- Matriz de Confusión:



Es una manera de vizualizar los valores predichos contra los valores reales, para probar el modelo de clasificación.


- Accuracy



Accuracy se encarga de decirnos cuál es el porcentaje de predicciones correctas, del total de predicciones hechas por el modelo, sin embargo, puede estar sesgado si el conjunto de datos esta desbalanceado y tiene mayor cantidad de datos de una clasificación comparada a la otra.

- Precision



Precision nos dice cuántas de las predicciones positivo eran correctas, lo cual nos sirve cuando queremos evitar falsos positivos o errores de tipo I. Qué tan correctas son mis predicciones positivas.

- Recall (Sensitivity)



Recall mide la capacidad del modelo para capturar los casos positivos, nos ayuda cuando nos interesa minimizar falsos negativos, como en los casos médicos cuando no hay problema por una falsa alarma pero si cuando algo pasa desapercibido.

- Specificity



Define la capacidad del modelo para correctamente identificar entradas negativas, es decir, de todos los verdaderos negativos, cuántos clasificó correctamente el modelo.


- F1




Es una buena idea de la combinación de precision y recall, cuando nos importan tanto la capacidad del modelo por capturar los casos positivos, como la cantidad de predicciones positivas hechas, que fueron correctas. 

- AUC-ROC

ROC (Receiver Operator Characteristic) es una curva de probabilidad que grafica la tasa de verdaderos positivos contra la tasa de falsos positivos, en distinto umbrales de probabilidad para un modelo de clasificación. El AUC (Area Under the Curve) es la forma en la que se mide la habilidad de un clasificador de distinguir entre clases, mientras más alto sea el AUC, mejor es el modelo, ya que cuando es igual a 1, el modelo es capaz de predecir perfectamente entre todos los positivos y negativos. 






[4][5]

### HIPERPARÁMETROS

Los hipérparámetros son variables de configuración que se utilizan en los modelos de aprendizaje estadístico, antes de generar los modelos para entrenar el proceso del mismo. Existen distintos tipos de hiperparámetros para cada modelo dado, y se buscan optimizar, ya que a diferencia de los parámetros ordinarios de un modelo, los cuales se calculan por el mismo modelo, los hiperparámetros deben ser supervisados para controlar cada modelo dado. 

Ejemplos de hiperparámetros son, gamma, C y el kernel para un modelo que utilice máquinas de soporte vectorial, o la cantidad de capas y de neuronas por capa de una red neuronal. En ambos casos, los hiperparámetros pueden ser configurados previo al entrenamiento del modelo, y a partir de un proceso de optimización, encontrar el hiperparámetro que mejor entrene el modelo de acuerdo a las necesidades del mismo. 

[6]

### OPTIMIZACIÓN BAYESIANA

El teorema de bayes, uno de los cimientos del estudio de la probabilidad, dice, entre muchas otras cosas, que conforme actualizamos la información de un evento, debemos actualizar nuestras creencias al mismo tiempo. Por ende, las probabilidades de dados eventos cambian conforme los eventos suceden. Aplicando este teorema a la optimización, tenemos la optimización bayesiana, que de lo que se encarga es de minimizar o maximizar un función en un domino dado, a través de un proceso gaussiano para la función a optimizar y un proceso de actualización para el mismo proceso gaussiano con cada nueva evaluación de la función.

En términos generales, lo que busca la optimización bayesiana es reducir el costo de evaluar un escenario de manera real o simulada, haciendo inferencia sobre la probabilidad del siguiente punto óptimo de una función, para luego únicamente evaluar ese punto y a partir de esa información nueva, actualizar las creencias y el proceso para repetir hasta que tengamos un posible máximo o mínimo para la función de interés. 

[7]

### OPTUNA

Optuna es es un algoritmo de optimización de hiperparámetros que utiliza optimización bayesiana pare encontrar los mejores valores de un conjunto de hiperparámetros para un modelo. Es fácil de integrar, más eficiente que otros métodos de búsqueda como Grid Search y es computacionalmente más ligero. Esto se debe a su enfoque bayesiano, que va actualizando el algoritmo en tiempo real, conforme va estimando mejores valores para los hiperparámetros. 

En lugar de utilizar procesos gaussianos, optuna utiliza un algoritmo llamado TPE (Tree-structured Parsen Estimator). Este algoritmo optimiza a partir de dos funciones de densidad de probabilidad, una para los valores de los hiperparámetros que van dando buenos resultados y otra para los que no.

```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelBinarizer

from sklearn.pipeline import Pipeline

from sklearn.compose import ColumnTransformer


from sklearn.model_selection import cross_val_score

from sklearn.neural_network import MLPClassifier

import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler

from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt

from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF

from sklearn.decomposition import PCA

import optuna
```

### Limpieza de datos y preprocesamiento

```python
# Leemos los datos y verificamos si hay valores nulos o nans que puedan malfuncionar el modelo
df = pd.read_csv("loan_data.csv")
df.head()
```

```python
# Definimos cuales variables vamos a modelar como numéricas y cuáles como categóricas

numerical_features = ["person_age", "person_income", "person_emp_exp", "loan_amnt", 
                      "loan_int_rate", "loan_percent_income", "cb_person_cred_hist_length",
                      "credit_score"]
categorical_features = ["person_gender", "person_education", "person_home_ownership", "loan_intent", "previous_loan_defaults_on_file" ]
```

```python
# Elegimos cuales datos serán nuestros predictores y cual variable sera la objetivo

x = df.copy()
x = x.drop(columns=["loan_status"])
y = df["loan_status"]
```

```python
# Definimos el mismo preprocesamiento para todos los pipelines que vayamos a construir

pca = PCA(n_components=0.95)

numerical_transformer = StandardScaler()
categorical_transformer = OneHotEncoder()

preprocessor = ColumnTransformer(
    transformers = [
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ]
)
```

### REGRESIÓN LOGÍSTICA CON PENALIZACIÓN

```python
model = LogisticRegression(penalty="l2", fit_intercept = True)

pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('pca', pca),  
    ('model', model)
])

pipeline.fit(x, y)
```

```python
y_score = pipeline.predict_proba(x)

lb = LabelBinarizer()
y_onehot = lb.fit_transform(y)
y_onehot = np.hstack([1 - y_onehot, y_onehot])
```

```python
import matplotlib.pyplot as plt

from sklearn.metrics import RocCurveDisplay

names = ["Default", "Not Default"]

classes = [0, 1]
for class_of_interest in classes:
    
    class_id = np.flatnonzero(lb.classes_ == class_of_interest)[0]
    display = RocCurveDisplay.from_predictions(
        y_onehot[:, class_id],
        y_score[:, class_id],
        name = f"{names[class_of_interest]} vs the rest",
        color = "darkgreen",
        plot_chance_level = True
    )
    _ = display.ax_.set(
        xlabel = "False Positive Rate",
        ylabel = "True Positive Rate",
        title = "One-vs-Rest curves: \n" + str(names[class_of_interest]) + " vs rest",
    )
```

```python
from sklearn.model_selection import cross_val_score

cv = 10
scores = cross_val_score(pipeline, x, y, scoring = "roc_auc_ovr", cv=cv)
auc = scores

auc_mean = auc.mean()
auc_std = auc.std()

print(auc_mean, auc_std)
```

#### OPTIMIZACIÓN DE HIPERPARÁMETROS

```python
# Definir función a optimizar
def objective(trial):
    # Sugerir parámetros
    n_components = trial.suggest_int("n_components", 5, min(len(numerical_features), 50))  
    C = trial.suggest_loguniform("C", 1e-4, 10)  

    pca = PCA(n_components=n_components)

    # Definir el modelo
    model = LogisticRegression(C=C, penalty="l2", fit_intercept=True, max_iter=1000)

    # Definir pipeline
    pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('pca', pca),
        ('model', model)
    ])

    # Realizar cv y obtener el score promedio
    score = cross_val_score(pipeline, x, y, cv=5, scoring="accuracy", n_jobs=-1).mean()
    
    return score  

#OPTUNA
study = optuna.create_study(direction="maximize")  # Maximize accuracy
study.optimize(objective, n_trials=50)  # Try 50 different sets of hyperparameters


print("Best hyperparameters:", study.best_params)
```

#### MODELO OPTIMIZADO (HIPERPARÁMETROS ÓPTIMOS)

```python
model = LogisticRegression(penalty="l2", fit_intercept=True, C=1.6914066390165852)

pca = PCA(n_components=8)

pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('pca', pca),  
    ('model', model)
])
```

```python
from sklearn.model_selection import cross_val_score

cv = 10
scores = cross_val_score(pipeline, x, y, scoring = "roc_auc_ovr", cv=cv)
auc = scores

auc_mean = auc.mean()
auc_std = auc.std()

print(auc_mean, auc_std)
```

#### CONCLUSIONES (REGRESIÓN LOGÍSTICA)

Este modelo, es el más sencillo de los tres que se van a tocar en este proyecto, computacionalmente es el más ligero, pero al mismo tiempo también puede ser el mas impreciso, ya que es muy simple. Sin embargo, vemos que el AUC del modelo original, así como el promedio del métrico luego de los K-Folds Validation es de 95%, lo cuál representa una precisión muy eficiente. 

Aún así, podemos buscar optimizar los hiperparámetros del modelo, que en esta caso son los componentes del PCA y C de la regresión, y al hacerlo e implementar sus valores en el modelo, vemos que los métricos son extremadamente similares, casi insignificativamente mejores, entonces no encontramos mucho valor en optimizar estos valores para este modelo en específico.

#### OPTIMIZACIÓN DE HIPERPARÁMETROS

```python
x_sampled, _, y_sampled, _ = train_test_split(x, y, train_size=20000, stratify=y, random_state=42)
```

```python
# Definir función a optimizar
def objective(trial):
    # Sugerir parámetros
    n_components = trial.suggest_int("n_components", 5, min(len(numerical_features), 50))  
    C = trial.suggest_loguniform("C", 1e-3, 1e2)  
    gamma = trial.suggest_loguniform("gamma", 1e-3, 1e1)  

    
    pca = PCA(n_components=n_components)

    # Definir el modelo
    model = SVC(probability=True, kernel="rbf", C=C, gamma=gamma)

    # Definir pipeline
    pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('pca', pca),
        ('model', model)
    ])

    # Realizar cv y obtener el score promedio
    score = cross_val_score(pipeline, x_sampled, y_sampled, cv=2, scoring="accuracy", n_jobs=-1).mean()
    
    return score  # Higher is better

#OPTUNA
study = optuna.create_study(direction="maximize")  # Maximize accuracy
study.optimize(objective, n_trials=20)  # Try 50 different sets of hyperparameters


print("Best hyperparameters:", study.best_params)
```

Luego de optimizar los hiperparámetros de este modelo, que una vez más fueron el número de componentes del PCA y C, más gamma para este modelo en específico, vimos un aumento muy ligero en el AUC promedio del cross-validation del modelo, aproximadamente unos 3 puntos base, de 95.74% a 95.77%, un incremento pequeño, pero quizás sustancial en esquemas más grandes. #### OPTIMIZACIÓN DE PARÁMETROS

```python
import optuna


# Definir función a optimizar
def objective(trial):
    # Sugerir parámetros
    n_components = trial.suggest_int("n_components", 5, min(len(numerical_features), 50))  
    hidden_layer_sizes = tuple(trial.suggest_int(f"layer_{i}", 10, 200) for i in range(trial.suggest_int("num_layers", 1, 3)))  
    alpha = trial.suggest_loguniform("alpha", 1e-5, 1e-1)  
    learning_rate_init = trial.suggest_loguniform("learning_rate_init", 1e-4, 1e-1)  

    pca = PCA(n_components=n_components)

    # Definir el modelo
    model = MLPClassifier(
        random_state=1,
        hidden_layer_sizes=hidden_layer_sizes,
        alpha=alpha,
        learning_rate_init=learning_rate_init,
        max_iter=1000
    )

    # Definir pipeline
    pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('pca', pca),
        ('model', model)
    ])

    # Realizar cv y obtener el score promedio
    score = cross_val_score(pipeline, x_sampled, y_sampled, cv=3, scoring="accuracy", n_jobs=-1).mean()
    
    return score  # Higher is better

#OPTUNA
study = optuna.create_study(direction="maximize")  # Maximize accuracy
study.optimize(objective, n_trials=20)  # Try 50 different sets of hyperparameters


print("Best hyperparameters:", study.best_params)
```

```python
from sklearn.model_selection import cross_val_score

cv = 5
scores = cross_val_score(pipeline, x, y, scoring = "roc_auc_ovr", cv=cv)
auc = scores

auc_mean = auc.mean()
auc_std = auc.std()

print(auc_mean, auc_std)
```
